```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from matplotlib import style
style.use("ggplot")
```

```python
N = 15
x = np.random.rand(N) * 3
y = x**3
colors = np.random.rand(N)
```

```python
area = np.pi * (15 * np.random.rand(N))**2  # 0 to 15 point radii
```

```python
fig = plt.scatter(x, y, s=100, c="green", alpha=1)
```

```python
# Create a Figure object.
fig = plt.figure()
# Create an Axes object.
ax = fig.add_subplot(1,1,1) # one row, one column, first plot
# Plot the data.
ax.scatter(x, y, s=50, color="green")
# Add a title.
ax.set_title("plot title, including Ω".decode("utf8"))
# Add some axis labels.
ax.set_xlabel("my x")
ax.set_ylabel("y")
```

```python
f, ax = plt.subplots(1, 1, figsize=(5,4))

x = np.linspace(0, 10, 100)
y = np.power(x, 2)
ax.plot(x, y, color = "b")
ax.set_xlim((1, 5))
ax.set_ylim((0, 30))
ax.set_xlabel('my x label')
ax.set_ylabel('my y label')
ax.set_title('plot title, including $\Omega$')

plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0, 10, 50)
y1 = np.power(x, 2)
y2 = np.power(x, 3)

plt.plot(x, y1, 'b-', label = '$x^2$')
plt.plot(x, y2, 'go', label = '$x^3$')

plt.xlim((1, 5))
plt.ylim((0, 30))

plt.xlabel('my x label')
plt.ylabel('my y label')
plt.title('plot title, including $\Omega$')

plt.legend()
plt.savefig('fig1.png')
#plt.show()
```

```python
data = np.random.randn(1000)
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(6, 3))
ax1.hist(data, bins=30, normed=True, color='b')
ax2.hist(data, bins=30, normed=True, color='r', cumulative=True)
plt.savefig('fig2.png')
#plt.show()
```

```python
samp1 = np.random.normal(loc=0., scale=1., size=100)
samp2 = np.random.normal(loc=1., scale=2., size=100)
samp3 = np.random.normal(loc=0.3, scale=1.2, size=100)

f, ax = plt.subplots(1, 1, figsize=(5,4))

ax.boxplot((samp1, samp2, samp3))
ax.set_xticklabels(['sample1', 'sample2', 'sample3'])
plt.savefig('fig3.png')
#plt.show()
```
